```python
df_sub = df[(df['sample_label'] == 'neu') | (df['sample_label'] == 'ang') | 
            (df['sample_label'] == 'sad') | (df['sample_label'] == 'hap')]
```

```python
import os
import glob
import shutil
```

```python
root_dir = './wav_data/pretrain/IEMOCAP_resample/'
output_dir = './wav_data/pretrain/IEMOCAP_sub'

for i in range(len(df_sub)):
    file_name = root_dir + df_sub['sample_name'][i] + '.wav'
    shutil.copy(file_name, output_dir)
```

```python
eval_ses = 'Ses0' + str(1)
train_idx = []
eval_idx = []

for i, sample_name in enumerate(df_sub['sample_name']):
    if eval_ses in sample_name:
        eval_idx.append(i)
    else:
        train_idx.append(i)
```

```python
for ses_num in range(1, 6):
    eval_ses = 'Ses0' + str(ses_num)
    train_index = []
    eval_index = []
    for i, sample_name in enumerate(df_sub['sample_name']):
        if eval_ses in sample_name:
            eval_index.append(i)
        else:
            train_index.append(i)
    print(len(train_index), len(eval_index), len(train_index)+len(eval_index))
```
