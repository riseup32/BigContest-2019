```python
def outlier_fill(data, days):
    data_copy = data.copy()
    for day in days:
        data_copy.loc[:, day] = data_copy.iloc[:, np.arange(1, 29) % 7 == (day % 7)].drop(day, axis = 1).mean(axis = 1)
    return data_copy
```

```python
def outlier_fill_2(data, days):
    data_copy = data.copy()
    data_copy.loc[:, days] = data_copy.iloc[:, np.arange(1, 29) % 7 == (days[0] % 7)].drop(days, axis = 1).mean(axis = 1)
    return data_copy
```

```python
def plot_show_leave(data1, data2, variable):
    plt.plot(range(1, 29), data1.groupby(['acc_id', 'day'])[variable].sum().unstack('day').fillna(0).mean(), label = 'reamin')
    plt.plot(range(1, 29), data2.groupby(['acc_id', 'day'])[variable].sum().unstack('day').fillna(0).mean(), label = 'leave')
    plt.title(variable, fontsize = 16)
    plt.xlabel('day', fontsize = 14)
    plt.ylabel('mean', fontsize = 14)
    plt.legend()
    plt.show()
```

```python
def plot_show_amount(data1, data2, variable):
    plt.plot(range(1, 29), data1.groupby(['acc_id', 'day'])[variable].sum().unstack('day').fillna(0).mean(), label = 'non_spent')
    plt.plot(range(1, 29), data2.groupby(['acc_id', 'day'])[variable].sum().unstack('day').fillna(0).mean(), label = 'spent')
    plt.title(variable, fontsize = 16)
    plt.xlabel('day', fontsize = 14)
    plt.ylabel('mean', fontsize = 14)
    plt.legend()
    plt.show()
```
